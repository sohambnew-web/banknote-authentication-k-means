# To determine whether the bank note is real or fake using K-means

```python
#First import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("Banknote-dataset.csv")    #Importing the csv data file
```

```python
#Lets transform this data into 'arrays'
A1 = data ["V1"]
A2 = data ["V2"]
```

```python
#Calculate the "mean" using numpy mean fucntion.
mean_A1  = np.mean(A1)
mean_A2 = np.mean(A2)
print(mean_A1, mean_A2)

#Calculate the "standard deviation" using numpy fucntion.
std_A1 = np.std(A1)
std_A2 = np.std(A2)
print(std_A1, std_A2)
```

```python
#Now coverting arrays A1 and A2 into 2D array A.
A = np.column_stack((A1, A2))
```

```python
#Now use Kmeans
km = KMeans(n_clusters = 3)   #These are number of clusters
km_res = km.fit(A)
clusters = km_res.cluster_centers_
```

```python
#Now plotting the graph
plt.scatter(A1, A2)
plt.scatter(clusters[:, 0], clusters[:, 1], s = 100)
plt.xlabel("V1")
plt.ylabel("V2")
plt.show()
```

```python
#Now try for different values of n_clusters

for k in [2, 3, 4, 5]:
    km = KMeans (n_clusters = k, random_state = 42)
    k_res = km.fit(A)
    clusters = km.cluster_centers_
    plt.scatter(A1, A2)
    plt.scatter(clusters[:, 0], clusters[:, 1], s = 100, c='red')
    plt.title (f"K-Means with k ={k}")
    plt.xlabel("V1")
    plt.ylabel("V2")
    plt.show()
```
